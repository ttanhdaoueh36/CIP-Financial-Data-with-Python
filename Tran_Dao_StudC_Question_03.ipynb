### Question 3: Do the S&P 500 members or the 500 most active non-members have the higher volume on average? (Activity is measured by the trading volume as of October 31st. )

- Step 1: Subsetting volume data for S&P 500 members and non-members on the last day of October.
- Step 2: Calculating the average **trading volume** for 2 groups mentioned above.
- Step 3: Comparing the 2 values of average **trading volume** and concluding which group has the higher volume on average.

##### Step 1: Subsetting volume data for S&P 500 members and non-members on the last day of October.

```python
# Importing the packages
import pandas as pd
import numpy as np
```

```python
# Read the dataset that contains the trading volume of S&P 500 members and non-members
df=pd.read_csv("../Tran_Dao_Data/Tran_Dao_StudC_combined_stock_data_stage.csv")
df
```

```python
df['S&P 500'].value_counts()
```

```python
# Subsetting data for tickers that are members of S&P 500 and non-members only
# It means not to select "SPDR S&P 500 Trust"
df = df[df['S&P 500'] != 'SPDR S&P 500 Trust']
df
```

```python
df['S&P 500'].value_counts()
```

##### Step 2: Calculating the average **trading volume** for 2 groups mentioned above.

```python
df['Date'].max()
```

```python
# We would like to calculate the average trading volume on the last date of October,
# First, we need the lastest records of 'Volume' on '2021-10-29' 
# so select rows that data are measured on '2021-10-29'
df_volume=df[df['Date'] == '2021-10-29']
df_volume
```

```python
# Calculating the trading volume and store the result in a new column 'Trading_Volume'
df_volume['Trading_Volume'] = df_volume['Close']*df_volume['Volume']
df_volume.sample(3)
```

**Group by** membership of S&P 500 and find the mean of trading volume with aggregation function:

```python
result=df_volume.groupby(['S&P 500']).agg({'Trading_Volume': ['mean']}).reset_index()
result
```

#####   Step 3: Comparing the 2 values of average **trading volume** and concluding which group has the higher volume on average.

```python
avg_sp500=result.iloc[0,1]
avg_sp500
```

```python
avg_non_sp500=result.iloc[1,1]
avg_non_sp500
```

```python
avg_sp500>avg_non_sp500
```

**Answer**: The returned result is **True** which means that the S&P 500 members have higher volume on average compared to the 500 most active non-members.
