```python
# Importing the packages
import pandas as pd
import numpy as np
```

```python
# Subsetting data for tickers that are members of S&P 500 only

is_member = df['S&P 500'] == 'member'
df_sp500 = df[is_member]
df_sp500.sample(3)
```

##### Step 2: Calculate the average performing S&P 500 stocks in 2021

```python
df_sp500.info()
```

```python
# Convert "Date" into datetime
df_sp500['Date']=pd.to_datetime(df_sp500['Date'])
df_sp500.dtypes
```

```python
df_sp500['Date'].max()
```

```python
df_sp500['Date'].min()
```

```python
# Selecting rows that have data on '2020-12-31' and '2021-10-29'
df_sp500=df_sp500[df_sp500['Date'].isin(['2020-12-31','2021-10-29'])]
df_sp500
```

```python
# Transform the long format into wide format:
df_sp500_wide=df_sp500.pivot(index=["Ticker","ESG Score"],columns="Date",values="Close")
df_sp500_wide
```

```python
# Checking if any missing values after transfering
df_sp500_wide.isna().sum()
```

```python
# Drop missing values
df_sp500_wide=df_sp500_wide.dropna()
df_sp500_wide
```

```python
# Calculating the pure performance and store the result in a new column 'Change'
df_sp500_wide['Change'] = (df_sp500_wide['2021-10-29']-df_sp500_wide['2020-12-31'])/df_sp500_wide['2020-12-31']
df_sp500_wide       
```

```python
# Average of pure performance is:
df_sp500_wide['Change'].mean()
```

```python
# Create new column titled 'Performance'
# then assign which ticker has higher than average 'Change' will be 'Outperform', 
# lower than average 'Change' will be 'Underperform'
df_sp500_wide['Performance'] = np.where(df_sp500_wide['Change']> df_sp500_wide['Change'].mean(),
                                                'Outperform', 
                                                'Underperform')

df_sp500_wide
```

##### Step 3: Find the best performing S&P 500 stocks in 2021, then save it as "Tran_Dao_StudC_top5_sp500_stage.csv" and send that file to student B for her work.

```python
# Rank the tickers based on their "Change" values
# Use the method 'dense' rank if found 2 values are same, then assign the same rank 
# Ranking of tickers in ascending order
# which means the higher "Change",the better rank
df_sp500_wide['Perform_Rank']=df_sp500_wide['Change'].rank(ascending=0,method='dense')
df_sp500_wide.head()
```

```python
# Sorting values of 'Perform_Rank' in ascending order
sp500_ranking=df_sp500_wide.sort_values(['Perform_Rank'],ascending=True)
sp500_ranking
```

```python
sp500_ranking['Perform_Rank']=sp500_ranking['Perform_Rank'].astype('int64')
sp500_ranking
```

```python
sp500_ranking.reset_index(inplace=True)
sp500_ranking.info()
```

```python
# Write the data into a csv file and then save it for Question 4
sp500_ranking.to_csv("../Tran_Dao_Data/Tran_Dao_StudC_sp500_ranking_stage.csv", index = False)
```

```python
# In order to answer the GROUP's Question 2, the top 5 tickers from rank 1st to 5th of Performance Rank
# would be provided to student B.

# Select the 5 best ticker from rank 1st to rank 5th
top5_sp500 = sp500_ranking.loc[sp500_ranking['Perform_Rank'] < 6,:]
top5_sp500
```

##### Step 5: Manipulate the data in order to answer the question 2

```python
top5_sp500_output.columns
```

```python
# Rename several columns
top5_sp500_output.rename(columns = {"reporting_person":"Person",
                                    "issuer":"Company",
                                    "issuer_tradingSymbol":"Ticker",                                        
                                    "transaction_aquireDisposedCode":"Transaction_type",
                                    "transaction_date":"Transaction_date",
                                    "transaction_shares":"Transaction_shares"
                                   }, 
                        inplace = True)
```

```python
# Subsetting the data for necessary attributes

top5_sp500_output = top5_sp500_output[['Person',
                                       'Company',
                                       'Ticker', 
                                       'Transaction_date',
                                       'Transaction_shares',
                                      'Transaction_type']]
top5_sp500_output
```

**Group by** person name,company, ticker name and their transaction type (A is for buyer, D is for seller) 
in order to check how many shares they buy or sell:

```python
transaction = top5_sp500_output.groupby(['Person','Company','Ticker','Transaction_type']).agg('sum').reset_index()
transaction
```

```python
# Transform the long format into wide format:
transaction_wide=transaction.pivot(index=['Person','Company','Ticker'],
                                        columns='Transaction_type',
                                        values='Transaction_shares')
```

```python
transaction_wide.head()
```

```python
transaction_wide.reset_index(inplace=True)
```

```python
transaction_wide.info()
```

```python
# Checking missing values
transaction_wide.isna().sum()
```

There are some missing values in column "A" means that this person have no acquired transactions.
Similarly, some missing values in column "D" means that this person have no disposed transactions.

Calculating the difference between acquired shares (column "A") and disposed shares(column "D")
in order to find out how much shares each person really exchanged in transaction.

```python
# First, convert "NaN" into "0":
transaction_full=transaction_wide.fillna(0)
transaction_full.head()
```

```python
transaction_full['A']=transaction_full['A'].astype('int64')
transaction_full['D']=transaction_full['D'].astype('int64')
```

```python
# Calculating the difference and store the result in a new column 'Difference'
transaction_full['Difference'] = transaction_full['A']-transaction_full['D']
transaction_full.head()
```

- If the value of "Difference" is positve, it means the corresponding person had acquired shares more than disposed shares. 

=> The largest buyer will be the person who has the highest value of 'Difference' between acquired and disposed.

- If the value of "Difference" is negative, it means the corresponding person had disposed shares more than acquired shares. 

=> The largest seller will be the person who has the lowest value of 'Difference' between acquired and disposed.

```python
# Sorting values of column 'Difference' in descending order in order to find the 3 biggest insiders
top_buyer=transaction_full.sort_values(['Difference'],ascending=False)
top_buyer.head(3)
```

```python
top_buyer_list=top_buyer.head(3)
top_buyer_list
```

```python
# Sorting values of column 'Difference' in ascending order in order to find the 3 biggest insiders
top_seller=transaction_full.sort_values(['Difference'],ascending=True)
top_seller.head(3)
```

```python
top_seller_list=top_seller.head(3)
top_seller_list
```
