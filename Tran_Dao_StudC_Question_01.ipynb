##### Step 1: Subsetting the data for necessary attributes 

```python
# Importing the packages
import pandas as pd
import numpy as np
```

```python
# Read the combined dataset 
df= pd.read_csv("../Tran_Dao_Data/Tran_Dao_StudC_combined_stock_data_stage.csv")
df.head()
```

```python
# Checking the uniques values of "Sector"
df.Sector.unique()
```

```python
# Subsetting data for tickers that are members of S&P 500 and belong to sector 'Information Technology' only

is_member = df['S&P 500'] == 'member'
is_IT = df['Sector']=='Information Technology'

df_sp500 = df[is_member & is_IT]
df_sp500.sample(3)
```

Because we focus on performance and ESG score of stocks so create a new dataframe which includes columns such as: 'Ticker', 'Date', 'Close', 'Sector' and 'ESG Score' to serve the purpose of the research question:

```python
# Subsetting data by columns
df_esg = df_sp500[['Ticker','Date','Close','Sector','ESG Score']]
df_esg.head()
```

##### Step 2: Calculate the average performing for S&P500 'Information Technology' (IT) stocks in 2021, evaluate which stocks outperformed the average of IT stocks performance.

To calculate the performance of a stock in 2021, we need the recored data on '2020-12-31' and the last date of October:

```python
df_esg['Date'].min()
```

```python
df_esg['Date'].max()
```

```python
# Selecting records that have data on '2020-12-31' and '2021-10-29'
df_esg=df_esg[df_esg['Date'].isin(['2020-12-31','2021-10-29'])]
df_esg.head()
```

```python
# Transform the long format into wide format:
df_esg_wide=df_esg.pivot(index=["Sector","Ticker","ESG Score"],columns="Date",values="Close")
df_esg_wide.reset_index()
```

```python
# Checking if any missing values after transfering
df_esg_wide.isna().sum()
```

```python
# Calculating the pure performance and store the result in a new column 'Change'
df_esg_wide['Change'] = (df_esg_wide['2021-10-29']-
                              df_esg_wide['2020-12-31'])/df_esg_wide['2020-12-31']
df_esg_wide.head()                           
```

```python
# Average of pure performance of all tickers is:
df_esg_wide['Change'].mean()
```

```python
# Create new column titled 'Performance'
# then assign which ticker has higher than average of 'Change' will be 'Outperform', 
# lower than average of'Change' will be 'Underperform':
df_esg_wide['Performance'] = np.where(
                                                df_esg_wide['Change']> df_esg_wide['Change'].mean(),
                                                'Outperform', 
                                                'Underperform')

df_esg_wide.head()
```

```python
# Selecting tickers that "Performance" is "Outperform"
is_outperform = df_esg_wide['Performance']=='Outperform'
df_esg_wide = df_esg_wide[is_outperform]
df_esg_wide.head()
```

```python
# Reset the index
df_esg_wide=df_esg_wide.reset_index()
```

##### Step 3: Find the 5 best ranked ESG stocks, then save it as "Tran_Dao_StudC_top5_esg_stage.csv" and send that file to student B for her work.

Rank the tickers based on their "ESG Score" and store the rank in a new column titled 'ESG_Rank'
which means the lowest ESG score corresponds to the best rank.

```python
df_esg_wide.columns
```

```python
# Rearrange the order of column names
df_esg_wide = df_esg_wide.reindex(columns=['Sector', 'Ticker', '2020-12-31', '2021-10-29', 'Change',
       'Performance', 'ESG Score'])
df_esg_wide.head(3)
```

```python
#  Rank the stock based on the ESG scores
#  Use the method 'dense' rank if found 2 values are same, then assign the same rank 
df_esg_wide['ESG_Rank']=df_esg_wide['ESG Score'].rank(ascending=1,method='dense')
df_esg_wide.head()
```

```python
# Sorting values of 'ESG_Rank' in ascending order
esg_ranking=df_esg_wide.sort_values(['ESG_Rank'],ascending=True,axis=0,ignore_index=True)
esg_ranking
```

```python
esg_ranking.groupby('ESG_Rank').agg('count')
```

```python
# In order to answer the GROUP's question number 1, the top 5 tickers from rank 1st to 5th of ESG Rank
# would be provided to student B.

# As we can see from the result above, there are:
# 2 tickers at the rank 1st, 
# 1 ticker at the rank 2nd, 
# 5 tickers at the rank 3rd.

# Therefore, we select:
# 2 tickers at the rank 1st, 
# 1 ticker at the rank 2nd,
# and 2 tickers at the rank 3rd which have the higher "Change" values than the other at the same rank 3rd. 
# Finally, we have only top 5 tickers with the best ESG score.
```

```python
# First, select the 8 best tickers at the rank 1st, 2nd and 3rd:
top8_esg = esg_ranking.loc[esg_ranking['ESG_Rank'] < 4,:]
top8_esg 
```

```python
# Then, select:
# 2 tickers at the rank 1st, 
# 1 ticker at the rank 2nd,
# and 2 tickers at the rank 3rd which have the higher "Change" values than the other at the same rank 3rd. 
# Finally, we have only top 5 tickers with the best ESG score.
# Those tickers in a list as below:
```

```python
is_top_5 = ['KEYS','CDW','HPQ','STX','ACN']
top5_esg = top8_esg[top8_esg['Ticker'].isin(is_top_5)]
```

```python
top5_esg.reset_index(drop=True, inplace=True)
top5_esg
```

##### Step 5: Manipulate the output data in order to answer the question 1


```python
top5_esg_output.columns
```

```python
# Rename several columns
top5_esg_output.rename(columns = {"reporting_person":"Person",
                                    "issuer":"Company",
                                    "issuer_tradingSymbol":"Ticker",                                        
                                    "transaction_aquireDisposedCode":"Transaction_type",
                                   "transaction_date":"Transaction_date",
                                    "transaction_code":"Transaction_code",
                                    "transaction_shares":"Transaction_shares"
                                   }, 
                        inplace = True)
```

```python
# Subsetting the data for necessary attributes

top5_esg_output = top5_esg_output[['Person',
                                       'Company',
                                       'Ticker', 
                                       'Transaction_date',
                                       'Transaction_shares',
                                      'Transaction_type']]
top5_esg_output
```

**Group by** person name,company, ticker name and their transaction type (A is for buyer, D is for seller) in order to check how many shares they buy or sell:

```python
transaction = top5_esg_output.groupby(['Person','Company','Ticker','Transaction_type']).agg('sum').reset_index()
transaction
```

```python
# Transform the long format into wide format:
transaction_wide=transaction.pivot(index=['Person','Company','Ticker'],
                                        columns='Transaction_type',
                                        values='Transaction_shares')
```

```python
transaction_wide.head()
```

```python
transaction_wide.reset_index(inplace=True)
```

```python
transaction_wide.info()
```

```python
# Checking missing values
transaction_wide.isna().sum()
```

There are some missing values in column "A" means that this person has no acquired transactions.
Similarly, some missing values in column "D" means that this person has no disposed transactions.

Calculating the difference between acquired shares (column "A") and disposed shares(column "D")
in order to find out how much shares each person really exchanged in transaction.

```python
# First, convert "NaN" into "0":
transaction_full=transaction_wide.fillna(0)
transaction_full.head()
```

```python
transaction_full['A']=transaction_full['A'].astype('int64')
transaction_full['D']=transaction_full['D'].astype('int64')
```

```python
# Calculating the difference and store the result in a new column 'Difference'
transaction_full['Difference'] = transaction_full['A']-transaction_full['D']
transaction_full.head()
```

- If the value of "Difference" is positve, it means the corresponding person had acquired shares more than disposed shares. 
- If the value of "Difference" is negative, it means the corresponding person had disposed shares more than acquired shares. 

In order to answer the research question, need to find out who has the highest positive 'Difference' between acquired and disposed.

```python
# Sorting values of column 'Difference' in descending order in order to find the 3 biggest insiders
top_buyer=transaction_full.sort_values(['Difference'],ascending=False)
top_buyer.head(3)
```

**Answer**: The 3 largest buyers in the five best ranking ESG Total Score of the above sector average performing S&P500 Technology stocks in 2021 are:

- Myers Marie	
- Nygaard Jeffrey D.
